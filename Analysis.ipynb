### Loading the workbook

```python
import pandas as pd

file_path = "FIF_Data_Analyst_Practical_Test_R2.xlsx"


```

```python
xls = pd.ExcelFile(file_path)
print(xls.sheet_names)
```

```python

dataset1 = pd.read_excel(file_path, sheet_name='Dataset 1')

dataset2 = pd.read_excel(file_path, sheet_name='Dataset 2')
```

```python

dataset1['CURRENT_BAL'] = pd.to_numeric(dataset1['CURRENT_BAL'], errors='coerce')
dataset1['DEPOSIT_ACCOUNT'] = pd.to_numeric(dataset1['DEPOSIT_ACCOUNT'], errors='coerce')

branch_summary = dataset1.groupby('BU_NO').agg({
    'CURRENT_BAL': 'sum',
    'DEPOSIT_ACCOUNT': 'sum'
}).reset_index()

branch_summary['DEPOSIT_TO_BAL_RATIO'] = branch_summary['DEPOSIT_ACCOUNT'] / branch_summary['CURRENT_BAL']

max_ratio_branch = branch_summary.loc[branch_summary['DEPOSIT_TO_BAL_RATIO'].idxmax()]
```

### QN1 :A report that shows the current outstanding balance & deposit account balance PER BRANCH

```python

pd.options.display.float_format = '{:,.2f}'.format


print(branch_summary)

print("\nBranch with highest deposit-to-balance ratio:")
print(max_ratio_branch)
```

```python
branch_summary.to_excel("branch_balances_report.xlsx", index=False)
```

### QN1(2): Branch with the highest ratio of deposits compared to outstanding portfolio          

```python

max_ratio_branch = branch_summary.loc[branch_summary['DEPOSIT_TO_BAL_RATIO'].idxmax()]

print("Branch with highest deposit-to-balance ratio:")
print(max_ratio_branch)
```

### QN2: Disbursements for the month of May 2019  of products which made the biggest contribution to overall sales.

```python



dataset1['DISBURSEMENT_DT'] = pd.to_datetime(dataset1['DISBURSEMENT_DT'], format='%m/%d/%Y')

may_2019_data = dataset1[(dataset1['DISBURSEMENT_DT'].dt.month == 5) & (dataset1['DISBURSEMENT_DT'].dt.year == 2019)]


product_contribution = may_2019_data.groupby('PROD_DESC')['DISBURSEMENT_AMT'].sum().reset_index()


total_disbursement = may_2019_data['DISBURSEMENT_AMT'].sum()


product_contribution['Percentage_Contribution'] = (product_contribution['DISBURSEMENT_AMT'] / total_disbursement) * 100


product_contribution = product_contribution.sort_values('DISBURSEMENT_AMT', ascending=False)


print(f"Total Disbursement for May 2019: {total_disbursement:,.2f}")
print("\nProducts and their contributions to total disbursements in May 2019:")
print(product_contribution)


product_contribution.to_excel("may_2019_product_contribution_with_percentage.xlsx", index=False)
```

### QN 4: clients with more than one loan

```python

merged_df = pd.merge(dataset1, dataset2, on='ACCT_NO')

client_loan_counts = merged_df.groupby('CUST_NO')['ACCT_NO'].count()

clients_with_multiple_loans = client_loan_counts[client_loan_counts > 1]


num_clients_with_multiple_loans = len(clients_with_multiple_loans)
print(f"Number of clients with more than one loan: {num_clients_with_multiple_loans}")
```

### QUESTION 5: How many Groups are there for Pamoja Plus Loans and how many Group Members are there?				


```python

pamoja_plus = dataset1[dataset1['PROD_DESC'] == 'Pamoja Plus Loan'].copy()


num_groups = pamoja_plus['GROUP_NAME'].nunique()


pamoja_plus['CNT_MALE'] = pd.to_numeric(pamoja_plus['CNT_MALE'], errors='coerce').fillna(0)
pamoja_plus['CNT_FEMALE'] = pd.to_numeric(pamoja_plus['CNT_FEMALE'], errors='coerce').fillna(0)


total_members = (pamoja_plus['CNT_MALE'] + pamoja_plus['CNT_FEMALE']).sum()

print(f"Number of groups for Pamoja Plus Loans: {num_groups}")
print(f"Total number of group members: {int(total_members)}")
```

### QUESTION 6:Managements need to have a summary of all products. You need to provide summary information that will allow them to answer the following questions:


```python

dataset1['DISBURSEMENT_AMT'] = pd.to_numeric(dataset1['DISBURSEMENT_AMT'], errors='coerce')
dataset1['CURRENT_BAL'] = pd.to_numeric(dataset1['CURRENT_BAL'], errors='coerce')
dataset1['Term_Month'] = pd.to_numeric(dataset1['Term_Month'], errors='coerce')


summary = dataset1.groupby('PROD_DESC').agg(
    Frequency=('ACCT_NO', 'count'),
    Total_Outstanding_Balance=('CURRENT_BAL', 'sum'),
    Avg_Term=('Term_Month', 'mean'),
    Min_Term=('Term_Month', 'min'),
    Max_Term=('Term_Month', 'max'),
    Avg_Disbursement=('DISBURSEMENT_AMT', 'mean'),
    Min_Disbursement=('DISBURSEMENT_AMT', 'min'),
    Max_Disbursement=('DISBURSEMENT_AMT', 'max'),
    Avg_Balance=('CURRENT_BAL', 'mean'),
    Min_Balance=('CURRENT_BAL', 'min'),
    Max_Balance=('CURRENT_BAL', 'max')
).reset_index()
```

```python
print(summary)
```

```python
summary.to_excel("product_summary_report.xlsx", index=False)
```
